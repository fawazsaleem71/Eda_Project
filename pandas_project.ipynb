##### Project focus on Retail Insights: A Comprehensive Sales Dataset. The information includes order details, customer details, product details, pricing, and shipping information, among other aspects of sales transactions. Here's a thorough analysis of every column:

#### Column Descriptions:

###### 1. Order No: Unique identifier for each order.
###### 2. Order Date: Date when the order was placed.
###### 3. Customer Name: Name of the customer placing the order.
###### 4. Address: Customer's address (one entry appears to be missing).
###### 5. City: City where the customer is located.
###### 6. State: State where the customer is located.
###### 7. Customer Type: Type of customer (e.g., retail, wholesale).
###### 8. Account Manager: Name of the account manager handling the order.
###### 9. Order Priority: Priority level of the order.
###### 10. Product Name: Name of the product being sold.
###### 11. Product Category: Category to which the product belongs.
###### 12. Product Container: Container type for the product.
###### 13. Ship Mode: Mode of shipping for the order.
###### 14. Ship Date: Date when the order was shipped.
###### 15. Cost Price: Cost price of the product.
###### 16. Retail Price: Retail price at which the product is sold.
###### 17. Profit Margin: Margin between retail and cost prices.
###### 18. Order Quantity: Quantity of products ordered.
###### 19. Sub Total: Subtotal cost of the order.
###### 20. Discount %: Percentage of discount applied to the order.
###### 21. Discount $: Dollar amount of the discount.
###### 22. Order Total: Total cost of the order after applying discounts.
###### 23. Shipping Cost: Cost associated with shipping the order.
###### 24. Total: Overall total cost, including product cost, discounts, and shipping.

#### 3. Rename and Replace Columns


```python
for i in df:
  j=i.replace(' ','_')
  df=df.rename(columns={i:j})
```

```python
df.info()
```

```python
for i in df[['Cost_Price','Retail_Price','Order_Total','Profit_Margin','Sub_Total','Discount_$','Shipping_Cost','Total']]:
    df[i]=df[i].str.replace('$','') #changing the value '$' into ' '
    df[i]=df[i].str.replace(',','') #changing the value ',' into ' '
    df[i]=df[i].astype('float') #changing the dtype as float
```

```python
df['Discount_%']=df['Discount_%'].str.replace('%','') #changing the value '%' into ' '
```

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'], format='%d-%m-%Y') #date format change into 'd-m-y'
```

```python
df['Ship_Date']=pd.to_datetime(df['Ship_Date'], format='%d-%m-%Y') #date format change into 'd-m-y'
```

##### 15- How does the shipping cost vary by order priority across different cities?

##### 18- identify the maximum delay time between ship date and order date?

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
df['Product_Name']=pd.DataFrame(df['Product_Name'])
# Calculate the difference in days
df['Shipping_Duration'] = (df['Ship_Date'] - df['Order_Date']).dt.days
df1=df[df['Shipping_Duration']>0]
# Display the result
print(df1[['Product_Name','Order_Date', 'Ship_Date','Shipping_Duration']].sort_values(by='Shipping_Duration',ascending=False).head())
```

##### The maximum delay time of five product is given above.
